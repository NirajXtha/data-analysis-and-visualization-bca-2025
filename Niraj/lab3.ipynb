# Lab 3: Different Types of data


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/AirIndia.csv')

df.columns = [col.strip().replace('\n','') for col in df.columns]
```

```python
nominal = df['Month']

plt.figure(figsize=(8,5))
nominal.value_counts().plot(kind='bar', color='orange')
plt.title('Nominal Data: Month - Bar Graph')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(6,6))
nominal.value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('deep'))
plt.title('Nominal Data: Month - Pie Chart')
plt.ylabel('')
plt.show()
```

```python
ordinal = df['FY']

ordinal_counts = ordinal.value_counts().sort_index()
plt.figure(figsize=(8,5))
ordinal_counts.plot(kind='bar', color='gold')
plt.title('Ordinal Data: Financial Year - Ordered Bar Graph')
plt.xlabel('Financial Year')
plt.ylabel('Count')
plt.show()
```

```python
interval = df['PAX. LOAD FACTOR#\r(IN %)'].astype(float)

plt.figure(figsize=(8,5))
plt.plot(df['Month'], interval, marker='o', linestyle='-', color='lightcoral')
plt.title('Interval Data: PAX LOAD FACTOR (%) - Line Chart')
plt.xlabel('Month')
plt.ylabel('Load Factor %')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x=df['Month'], y=interval, palette='muted')
plt.title('Interval Data: PAX LOAD FACTOR (%) - Bar Chart')
plt.xlabel('Month')
plt.ylabel('Load Factor %')
plt.xticks(rotation=45)
plt.show()
```

```python
ratio_cols = ['DEPARTURES', 'HOURS', 'KILOMETER\r(TH)', 'PASSENGERS CARRIED',
              'PASSENGER KMS. PERFORMED\r(TH)', 'AVAILABLE SEAT KILOMETRE\r(TH)']

for col in ratio_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

plt.figure(figsize=(10,6))
sns.scatterplot(x='Month', y='PASSENGERS CARRIED', data=df, color='blue', s=100)
plt.title('Ratio Data: Passengers Carried - Scatter Plot')
plt.xlabel('Month')
plt.ylabel('Passengers Carried')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x='Month', y='DEPARTURES', data=df, palette='Set2')
plt.title('Ratio Data: Departures - Bar Chart')
plt.xlabel('Month')
plt.ylabel('Number of Departures')
plt.xticks(rotation=45)
plt.show()

sns.lmplot(x='PASSENGERS CARRIED', y='PASSENGER KMS. PERFORMED\r(TH)', data=df, aspect=1.5)
plt.title('Ratio Data: Passengers Carried vs Passenger KMS - Lollipop/Slope Chart')
plt.xlabel('Passengers Carried')
plt.ylabel('Passenger KMS Performed (TH)')
plt.show()
```
